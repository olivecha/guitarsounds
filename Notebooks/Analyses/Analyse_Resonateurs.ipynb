```python
import matplotlib.pyplot as plt
import numpy as np
```

## Import the soundhole testing dataset

```python
# Filenames
fileOG = 'soundfiles/test_reso/originale.wav_mono2.wav'
fileres1 = 'soundfiles/test_reso/res1.wav_mono2.wav'
fileres3 = 'soundfiles/test_reso/res3.wav_mono2.wav'
fileres4 = 'soundfiles/test_reso/res4.wav_mono2.wav'
files = [fileOG, fileres1, fileres3, fileres4]
names = ['original', 'reso 1', 'reso 3', 'reso 4']

sounds = []
for file, name in zip(files, names):
    sounds.append(Sound(file, name=name).condition())
```

## Find the hemlotz resonator frequency

### Plots of the frequency peaks

```python
fig, axs = plt.subplots(2,2, figsize=(12,12))
axs = axs.reshape(1,-1)[0]
for sound, ax in zip(sounds, axs):
    plt.sca(ax)
    sound.SP.change('fft_range', 500)
    sound.signal.plot('peaks')
    ax.set_title(sound.name)
    ax.grid('on')
plt.tight_layout()
```

**Puisque la fondamentale est le E2 (86 Hz) la frquence du rsonnateur est cache par la fondamentale.**

## Analyse de la puissance spectrale dans la zone du rsonnateur

```python
fig, axs = plt.subplots(2,2, figsize=(12,12))
axs = axs.reshape(1,-1)[0]
for sound, ax in zip(sounds, axs):
    plt.sca(ax)
    sound.SP.change('fft_range', 140)
    sound.signal.plot('fft', c='k')
    ax.set_title(sound.name)
    ax.grid('on')
    ax.set_xlim((0,140))
    ax.set_ylim((10e-4, 10e-1))
plt.tight_layout()
```

- **Moins de puissance dans la frquence du rsonnateur pour le rsonateur 3, les autres ne sont pas significativement diffrents**
