## Creating a Multiple SoundPack

```python
file1 = 'soundfiles/flax_carbon/Wood_A0_1.wav'
file2 = 'soundfiles/flax_carbon/Carbon_D0_2.wav'
file3 = 'soundfiles/flax_carbon/Carbon_E0_1.wav'
file4 = 'soundfiles/flax_carbon/Wood_D0_2.wav'
file5 = 'soundfiles/flax_carbon/Wood_E0_1.wav'
file6 = 'soundfiles/flax_carbon/Wood_E1_1.wav'
files = [file1, file2, file3, file4, file5, file6]
names = ['Wood_A', 'Carbon_D', 'Carbon_E', 'Wood_D', 'Wood_E', 'Wood_E2']

MultPack = SoundPack(files, names=names)
```

## Creating a Dual SoundPack

```python
Wood = Sound('soundfiles/Wood_Guitar/Wood_D0_1.wav')
Wood.condition()
Carbon = Sound('soundfiles/flax_carbon/Carbon_D0_2.wav')
Carbon.condition()

DualPack = SoundPack(Wood, Carbon, names=['Wood', 'Carbon'])
```

```python
DualPack.compare_peaks()
```

```python
DualPack.fft_mirror()
```

```python
DualPack.fft_diff(ticks='bins')
```
