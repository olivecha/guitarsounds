```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Signal(object):
    
    def __init__(self, x):
        self.x = x
        self.plot = self.PlotClass(self)
        
    class PlotClass(object):
        
        
        
        def __init__(self, signal):
            self.parent = signal
            
        def p_1(self):
            plt.plot(self.parent.x, self.parent.x**2)
            plt.show()
            
        def p_2(self):
            plt.plot(self.parent.x, 3 ** self.parent.x)
            plt.show()
    
    def plot_function(self, n):
        
        if n == 1:
            self.plot.p_1()

        elif n == 2:
            self.plot.p_2()
            
        
```

```python
signal = Signal(np.linspace(1,12,100))
```

```python
signal.plot_function(1)
```

```python
signal.plot.p_1()
```
